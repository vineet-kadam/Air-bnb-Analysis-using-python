```python
indian_states = [
    'Andhra Pradesh','Arunachal Pradesh','Assam','Bihar','Chhattisgarh',
    'Goa','Gujarat','Haryana','Himachal Pradesh','Jharkhand','Karnataka',
    'Kerala','Madhya Pradesh','Maharashtra','Manipur','Meghalaya','Mizoram',
    'Nagaland','Odisha','Punjab','Rajasthan','Sikkim','Tamil Nadu',
    'Telangana','Tripura','Uttar Pradesh','Uttarakhand','West Bengal',
    'Delhi','Jammu And Kashmir','Ladakh','Puducherry'
]
df[['Part1', 'Part2']] = df['address'].str.rsplit(",", n=1, expand=True)

df['Part1'] = df['Part1'].str.strip().str.title()
df['Part2'] = df['Part2'].str.strip().str.title()

df['City'] = df['Part1']
df['State'] = df['Part2']

df.loc[df['Part1'].isin(indian_states), 'City'] = None
df.loc[df['Part1'].isin(indian_states), 'State'] = df['Part1']

df.drop(['Part1', 'Part2'], axis=1, inplace=True)
```
